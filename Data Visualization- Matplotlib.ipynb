Matplotlib ia a python library used for data visulization

You can creat bar-plots, scatter-plots, histograms and a lot more with matplotlib.

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x= np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
# Line plot

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color="green", linewidth=2,linestyle=':')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
y1=2*x
y2=3*x

y1,y2
```

```python
plt.plot(x,y1,color='orange',linewidth=2,linestyle=':')
plt.plot(x,y2,color='blue',linewidth=2,linestyle=':')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# bar plot
```

```python
student = {'Sujit':60,"Ajit":90,"Avinash":80}
```

```python
names= list(student.keys())
marks= list(student.values())
```

```python
names,marks
```

```python
plt.bar(names,marks,color="yellow")
plt.title("Marks of Students")
plt.xlabel("Names of Student")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# Creat above bar as a horizontal 

#   plt.barh()
```

```python
plt.barh(names,marks,color="yellow")
plt.title("Marks of Students")
plt.xlabel("Names of Student")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# Scatter-plot
```

```python
x = [90,10,20,40,30,50,40]
y = [10,20,30,4,50,60,20]

plt.scatter(x,y)
plt.grid(True)
```

```python
# two point on same gragh

x = [4,7,3,9,1,6]
y1 = [9,1,2,6,4,9]

y2= [4,7,8,1,2,6]

plt.scatter(x,y1,color="blue")
plt.scatter(x,y2,color="red")
plt.grid(True)
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
plt.hist(iris["Sepal.Length"],bins=20)
plt.show()
```

histogram is uses for understand the distribution of contionous variable

```python
plt.hist(iris["Petal.Width"],bins=20)
plt.show()
```
